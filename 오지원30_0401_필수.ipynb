# 타이타닉 데이터 전처리 - 2차

```python
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
df
```

- 결측치 확인

```python
df.isna().sum() 
```

- 중복된 컬럼 삭제

```python
df = df.drop(['class', 'embark_town', 'alive'], axis=1)  
df
```

```python
df = df.drop(['adult_male','fare'], axis=1)  #데이터 분석과정에서 중요도가 낮을 것 같은 컬럼 제거
df
```

- 결측치 처리

```python
df.isna().sum()  #결측치 개수 확인
```

```python
df = df.drop(['deck'], axis=1)  #결측치가 많은 deck열 제거
df
```

```python
df.embarked.value_counts()
```

```python
df = df.dropna(subset=['embarked'])  #embarked컬럼에 결측치가 존재하는 인덱스 제거 (2)
df
```

```python
df.groupby('pclass').agg(mean_age = ('age', 'mean'))
```

```python
df.groupby('sex').agg(mean2_age = ('age', 'mean'))
```

```python
df.groupby('alone').agg(mean3_age = ('age', 'mean'))  #두 평균의 차이가 큰 alone컬럼의 각 평균으로 결측치를 채운다.
```

```python
df['age']=df['age'].fillna(df.groupby('alone')['age'].transform('mean'))
df
```

```python
df.isna().sum()  # 결측치 모두 처리 완료
```

- 인코딩

```python
import pandas as pd
import numpy as np
```

```python
encoding_df = pd.get_dummies(df, columns=['sex', 'embarked', 'who', 'alone'], prefix=['sex', 'embarked', 'who', 'alone'])
```

```python
encoding_df #원핫 인코딩한 데이터프레임
```

```python
encoding_df.info()
```

```python
encoding_df.to_csv('오지원30_0401.csv')
```

- 데이터 나누기

```python
x_df = encoding_df.drop('survived',axis=1)
y_df = encoding_df['survived'] 
```

```python
encoding_df
```

```python
x_df
```

```python
y_df
```

- train, test로 나눔

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_df, y_df, test_size=0.1, random_state=111)
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
y_test
```

- 정확도 확인

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
dt_clf = DecisionTreeClassifier(random_state=111)
rf_clf = RandomForestClassifier(random_state=111)
lr_clf = LogisticRegression(solver ='liblinear')
```

```python
dt_clf.fit(x_train, y_train)
dt_pred = dt_clf.predict(x_test)
print('DT정확도:{0:.4f}'.format(accuracy_score(y_test, dt_pred)))

rf_clf.fit(x_train, y_train)
rf_pred = rf_clf.predict(x_test)
print('RF정확도:{0:.4f}'.format(accuracy_score(y_test, rf_pred)))

lr_clf.fit(x_train, y_train)
lr_pred = lr_clf.predict(x_test)
print('LR정확도:{0:.4f}'.format(accuracy_score(y_test, lr_pred)))
```
