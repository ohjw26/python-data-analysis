- 시리즈 형태로 반복문 이용

```python
#mpg 데이터 중에서 연료의 종류가 p인 데이터를 출력하여 제조사별 카운팅
def get_fl_spe(series):
    fl_cnt =len([ _ for x in series if x[0] in 'p'])
    return fl_cnt
```

```python
mpg.groupby('manufacturer').agg({'fl':get_fl_spe}).reset_index() #제조사별
```

```python
#mpg 데이터 중에서 모델이 a4인 데이터를 출력하여 제조사별 카운팅
def get_model_spe(series):
    model_cnt =len([ _ for x in series if x[0] in 'a4'])
    return model_cnt
```

```python
mpg.groupby('manufacturer').agg({'model':get_model_spe}).reset_index()
```

```python
# 제조사 별로 제조년도가 2000년 이상인것과 2000년 미만인 것으로 나누기
def get_m_year(series):
    if mpg.iloc[series.index[0], : ]['manufacturer']=='audi':
        year_cnt = len([_ for x in series if x>=2000])
    else:
        year_cnt = len([_ for x in series if x<2000])
    return year_cnt
```

```python
mpg.groupby('manufacturer').agg({'year':get_m_year}).reset_index()
```
