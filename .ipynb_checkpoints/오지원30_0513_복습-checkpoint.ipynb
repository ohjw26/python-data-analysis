## 반복문 함수

* iterrows

```python
#idx:인덱스, row:전체 데이터프레임
for idx, row in mpg.iterrows():
    print(idx,row)
```

```python
for idx, row in mpg.iterrows():
    print(idx,row['hwy'])
```

```python
for idx, row in mpg.iterrows():
    print(idx,row['model'])
```

* itertuples

```python
#첫필드:인덱스, 나머지열
for row in mpg.itertuples():
    print(row)
```

```python
for row in mpg.itertuples():
    print(row.manufacturer, row.cty)
```

```python
for col in mpg.columns:
    df_coltp = str(mpg[col].dtype)
    if df_coltp =='object':
        print(col,'object')
    else:
        print(col,'else')
```

```python
for col in mpg.columns:
    df_coltp = str(mpg[col].dtype)
    if df_coltp =='object':
        print(col, mpg[col].value_counts()) #값의 통계
    else:
        print(col,mpg[col].describe())
```

- 시리즈 형태로 반복문 이용하여 조건문

```python
#mpg 데이터 중 trans 변수에 m5 값이 있는 데이터를 출력하여 제조사별 카운팅
def get_trans_spe(series):
    tr_cnt =len([ _ for x in series if x[0] in 'm5'])
    return tr_cnt
```

```python
#제조사 별로 카운팅
mpg.groupby('manufacturer').agg({'trans':get_trans_spe}).reset_index()
```

```python
#cty평균 조건으로 데이터 묶기
def get_y_cty(series):
    if mpg.iloc[series.index[0], : ]['year'] == 2008:
        cty_cnt = len([_ for x in series if x>=16]) #cty평균 16 이상
    else:
        cty_cnt = len([_ for x in series if x<16]) #cty평균 16 미만
    return cty_cnt
```

```python
#연도별로 묶기
mpg.groupby('year').agg({'cty':get_y_cty}).reset_index()
```
