## 반복문을  통한 데이터 프레임 접근하기 
- 2가지 원리를 이해하자
- 1. 반복문에 들어가는 값 , 2. 반복문을 통해서 나오는 값 


```python
for i in range(10):
    print(i)
```

- range는 range(10)이라는 숫자가 나온다.

- in 뒤에 range와 시퀀스객체에 따라 값이 달라진다.

```python
a = ['사과','딸기','바나나','포도']
```

```python
for i in a:
    print(i)
```

- 시퀀스 객체 등으로 range가 아닌 다른 것으로 끝나면 시퀀스객체 안에 값을 가지고 온다.


- 시퀀스객체를 range로는 표현 못하나?

```python
for i in range(len(a)):
    print(a[i])
```

### 반복문 제공되는 함수들

- 시리즈 형태로 반복문을 이용해서 조건문을 더욱 정교하게 만들어보자!

- mpg 데이터 중에서 trans 변수에 m5 값이 있는 데이터를 출력하여 제조사별 카운팅을 하고 싶다!

- m5만 추출하는 함수를 만들어 보자!

```python
mpg
```

```python
def get_trans_spe(series):
    tr_cnt =len([ _ for x in series if x[0] in 'm6'])
    return tr_cnt
    
```

- 제조사별로 카운팅을 해야한다!

```python
get_trans_spe(mpg.trans)
```

```python
mpg.groupby('manufacturer').agg({'trans':get_trans_spe}).reset_index()
```

- 연도별로 cty 평균이 16미만 친구들과 16이상이 친구들을 묶어보자!

```python
def get_y_cty(series):
    if mpg.iloc[series.index[0], : ]['year']==2008:
        cty_cnt = len([_ for x in series if x>=16])
    else:
        cty_cnt = len([_ for x in series if x<16])
    return cty_cnt
    
```

```python
mpg.groupby('year').agg({'cty':get_y_cty}).reset_index()
```
