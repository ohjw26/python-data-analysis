# 데이터 프레임 전처리

* pandas를 이용하여 전처리
* 데이터 가공시 다양한 함수 사용 가능
* 문자열, 시계열, 수치형 모두 전처리 가능

* 데이터프레임은 행과 열로 접근

```python
import pandas as pd
```

```python
df = pd.read_csv('mpg.csv')
```

```python
df
```

## query()

* 원하는 데이터 추출

<문자열 데이터 조건>

```python
df.query("manufacturer == 'audi'").hwy
```

```python
df.query("manufacturer != 'toyota'").cty
```

<수치형 데이터 조건>

```python
df.query("year >= 2000")    #대소비교 가능
```

```python
df.query("displ <= 2.5")
```

```python
df.query("hwy >= 30 & year >= 2000")   #논리연산자 이용하여 2개 이상의 조건 만들기 가능
```

```python
df.query("cyl >= 6 or cty <= 17")
```

## 정렬 

* 기본값은 오름차순

<오름차순>

```python
df[['year', 'cyl']].sort_values('year')
```

```python
df[['year', 'cyl']].sort_values('cyl')
```

<내림차순>

```python
df[['year', 'cyl']].sort_values('cyl', ascending=False)
```

```python
df.query("manufacturer == 'hyundai'")[['year', 'cyl']].sort_values('cyl', ascending=False)
```

<두 개 이상 정렬>

```python
df[['displ','cty']].sort_values(['displ','cty'], ascending=[True,False])
#순서대로 displ은 오름차순, cty는 내림차순
```

## assign()

* 기존 데이터에 파생변수 추가

```python
import numpy as np
```

```python
df.assign(state = np.where(df['displ'] >= 3.0 , 'good', 'bad'))
#whrer함수는 해당 값의 위치를 찾아줌
#넘파이에서 제공
```

df.assign(state = np.where(df['displ'] >= 3.0 and df['year'] >= 2008, 'good', 'bad')) 

```python
df = df.assign(state2 = lambda x :(x['displ']+ x['year'])) 
```

```python
df
```

```python
np.where(df['state2']>= 2010)
```

## Groupby

* groupby(기분으로 묶을 컬럼/인덱스), agg(출력할 변수, 내가 볼 통계치)

```python
df
```

```python
df.groupby('manufacturer').agg(mean_year = ('year', 'mean'),
                       sum_year = ('year', 'sum'),
                       count_year = ('year', 'count'))
```

```python
df.groupby('model').agg(mean_year = ('year', 'mean'),
                       sum_year = ('year', 'sum'),
                       count_year = ('year', 'count'))
```

```python
df_2 = df.query("manufacturer == 'hyundai'").groupby('year').agg(mean_year = ('year', 'mean'),
                       sum_year = ('year', 'sum'),
                       count_year = ('year', 'count'))
```

```python
df_2
```

## 인덱스 리셋

```python
df_2 = df_2.reset_index()
```
