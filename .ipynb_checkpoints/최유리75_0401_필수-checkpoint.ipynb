#### 타이타닉 데이터 컬럼 설명
- survived : 0=사망, 1=생존
- pclass : 1=1등석, 2=2등석, 3=3등석
- sex: male=남성, female=여성
- age : 나이
- sibsp : 함께 탑승한 형제 및 배우자 수
- parch : 함께 탑승한 자녀 및 부모 수
- fare : 요금
- embarked : 탑승지 이름 앞글자, C=Cherbourg, Q=Queenstown, S=Southampton
- class : 객실 등급, First=1등급, Second=2등급, Third=3등급
- who : 남자=man, 여자=woman, 아이=child
- adult_male : 성인 남성인지 여부. True면 성인 남성, False면 그 외
- deck : 선실 번호 첫 알파벳, A,B,C,D,E,F,G
- embark_town : 탑승지 이름, Cherbourg, Queenstown, Southampton
- alive : no=사망, yes=생존
- alone : True=혼자탑승, False=가족과 함께 탑승

- 저번주 전처리 진행할 때, 모든 변수를 살려두는 방법으로 진행했었음.
- 그 파일로 생존자 예측해보기를 진행했을 때, DT정확도:1.0000, RF정확도:1.0000, LR정확도:1.0000 나와서 아쉬움이 남음.
- 이번에는 성격이 유사한 변수들은 제거할 예정
- survived 변수와 alive 변수는 같은 의미를 나타내므로 alive 변수 제거
- embarked, embark_town 변수는 같은 의미를 나타내므로 embark_town 변수 제거
- class : 객실 등급, First=1, Second=2, Third=3으로 대체할 예정
- sex: male은 0으로, female은 1로 인코딩할 예정
- 일단 계획은 이러하고, 필요하다면 추가적인 과정을 진행할 예정

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = sns.load_dataset('titanic')
```

```python
df.shape
```

```python
df.isna().sum()
```

- deck변수에는 결측 많으니 변수 자체를 제거 

```python
df = df.drop('deck' , axis = 1)
df
```

- survived와 alive는 같은 의미를 나타내므로 alive 변수 제거

```python
df = df.drop('alive' , axis = 1)
df
```

- embarked, embark_town은 같은 의미를 나타내므로 embark_town 변수 제거

```python
df = df.drop('embark_town' , axis = 1)
df
```

```python
df.isna().sum()
```

```python
df = df.dropna(subset=['embarked']) # embarked에 결측 있으면 그 행 제거
```

```python
df
```

- age의 결측치 채우기 : 중앙값으로 채우기

```python
df.age.median() # 중앙값
```

```python
df['age']= df['age'].fillna(28.0)
```

```python
df.isna().sum()
```

- class : 객실 등급, First=1, Second=2, Third=3으로 대체하기

```python
df['class'] = df['class'].replace({'First': 1, 'Second': 2, 'Third': 3})
```

```python
df
```

- sex: male은 0으로, female은 1로 인코딩

```python
df['sex'] = df['sex'].replace(['male', 'female'], [0, 1])
```

```python
df
```

```python
df['adult_male'] = df['adult_male'].replace({True: 1, False: 0}) # adult_male에서 True는 1, False는 0으로 인코딩
```

```python
df['alone'] = df['adult_male'].replace({True: 1, False: 0}) # adult_male에서 True는 1, False는 0으로 인코딩
```

```python
df
```

```python
df = pd.get_dummies(df, columns=['embarked'])
df
# embarked 변수를 원핫인코딩 진행시킴
# embarked 변수는 "C", "Q", "S"를 나타내는 세 개의 새로운 변수("embarked_C", "embarked_Q", "embarked_S")로 대체됨.
```

```python
df = pd.get_dummies(df, columns=['who'])
df
# who 변수를 원핫인코딩 진행시킴
# who 변수는 "child", "man", "woman"을 나타내는 세 개의 새로운 변수("who_child", "who_man", "who_woman")로 대체됨.
```

```python
df.isna().sum() # 각 변수별 결측값 개수 확인
```

```python
df.shape #(행,열) 개수 추출
```

```python
df
```

```python
df = df.select_dtypes(include = [np.number])
df #수치형 데이터만 추출하기
```

```python
x_df = df.drop('survived',axis=1) # 생존 열을 제외하기
y_df = df['survived'] # 생존은 정답 데이터. 정답은 따로 빼서 일단 나눈다. 
```

```python
x_df
```

```python
y_df
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x_df, y_df, test_size=0.2, random_state=111)
# x_df와 y_df 데이터셋을 train set과 test set으로 분할
# test set의 비율 0.2 -> 전체 데이터셋 중 30%를 test set으로 사용
```

```python
X_train
```

```python
y_train
```

```python
X_test
```

```python
y_test
```

### ML 알고리즘으로 타이타닉 생존자 예측해보기
- 의사결정나무
- 랜덤포레스트
- 로지스틱회귀

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# 결정트리, 랜덤포레스트, 로지스틱회귀 만들기
dt_clf = DecisionTreeClassifier(random_state=111)
rf_clf = RandomForestClassifier(random_state=111)
lr_clf = LogisticRegression(solver ='liblinear')
```

```python
dt_clf.fit(X_train, y_train)
dt_pred = dt_clf.predict(X_test)
print('DT정확도:{0:.4f}'.format(accuracy_score(y_test, dt_pred)))

rf_clf.fit(X_train, y_train)
rf_pred = rf_clf.predict(X_test)
print('RF정확도:{0:.4f}'.format(accuracy_score(y_test, rf_pred)))

lr_clf.fit(X_train, y_train)
lr_pred = lr_clf.predict(X_test)
print('LR정확도:{0:.4f}'.format(accuracy_score(y_test, lr_pred)))
```

```python
dt_pred #머신이 학습해서 만든 값 
```

```python
y_test #실제값
```

- 저번주 전처리 진행한 파일로 생존자 예측해보기를 진행 -> DT정확도:1.0000, RF정확도:1.0000, LR정확도:1.0000
- 새롭게 전처리 진행한 df로 생존자 예측해보기를 진행 -> DT정확도:0.7809, RF정확도:0.8202, LR정확도:0.8539
- 처음으로 ML 알고리즘을 이용해 예측해보기를 배웠기 때문에 내가 생각한 이유가 정확한지는 모르겠지만, 저번주 파일은 결측치를 최빈값, 중앙값으로 채워 처리하고, 인코딩 진행하고, 같은 성격을 갖는 변수를 제외하지 않고 모든 변수를 넣고 돌렸기 때문에 정확도가 모두 1.0000이 나온 것 같다고 생각이 들었음
- 이번에는 같은 성격을 보이는 변수를 제외하고, 결측치가 많은 열은 열 자체를 삭제하는 등 이전과 다른 전처리 방법들을 이용해봐서 모두 1.0000이 나오지 않고, 전과 다르게 DT정확도:0.7809, RF정확도:0.8202, LR정확도:0.8539 나온 것 같음.
