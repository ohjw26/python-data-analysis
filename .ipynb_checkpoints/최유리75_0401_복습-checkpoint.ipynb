```python
x_df = df.drop('survived',axis=1) # 생존 열을 제외하기
y_df = df['survived'] # 생존은 정답 데이터. 정답은 따로 빼서 일단 나눈다. 
```

- 모델에게 학습을 하기 위해서는 2가지 또는 3가지로 나눈다.
- 무엇을 나누는가? 데이터를 나눔
- 왜 데이터를 나누는 것인가?

- 중간고사 보는데 중간고사 시험지에 나올 문제를 그대로 공부 -> 시험 보면 거의 대부분 100점. 문제를 외울 것이기 때문 

- 머신한테도 모든 데이터를 다 주면 그 전체 데이터의 특성을 다 배울 것 -> 전체 데이터를 다주면 신뢰할 수 없음

- 데이터를 나눠서 준다는 개념
- 모델이 실제로 학습했을 때 성능이 얼마나 좋은지를 평가하기 위해서 나눈다.
- train, test로.

- train은 모델에게 모의고사처럼 학습을 시키는것. 중간고사에 나올만한 것을 배운다. 

- 실제 중간고사는 test로 본다. test데이터는 실제로 머신이 본 적이 없다. 어떤 패턴인지 모른다. 학습을 한 적이 없음.
- train으로 학습한 것 -> test를 이용해서 실제 pred값을 만들고 그 예측값과 실제 값이랑 비교를해서 성능을 평가한다.

### train, test 나누기

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x_df, y_df, test_size=0.2, random_state=111)
# x_df와 y_df 데이터셋을 train set과 test set으로 분할하는 코드
# test set의 비율 0.2 -> 전체 데이터셋 중 20%를 test set으로 사용
# random_state 파라미터는 데이터셋을 무작위로 분할할 때 사용되는 시드(seed)값을 설정
# 시드값을 설정 -> 같은 데이터셋을 사용할 때마다 동일한 결과를 얻을 수 있음
```

### ML 알고리즘으로 타이타닉 생존자 예측해보기!
- 의사결정나무
- 랜덤포레스트
- 로지스틱회귀

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# 결정트리, 랜덤포레스트, 로지스틱회귀 만들기
dt_clf = DecisionTreeClassifier(random_state=111)
rf_clf = RandomForestClassifier(random_state=111)
lr_clf = LogisticRegression(solver ='liblinear')
```

```python
dt_clf.fit(X_train, y_train)
dt_pred = dt_clf.predict(X_test)
print('DT정확도:{0:.4f}'.format(accuracy_score(y_test, dt_pred)))

rf_clf.fit(X_train, y_train)
rf_pred = rf_clf.predict(X_test)
print('RF정확도:{0:.4f}'.format(accuracy_score(y_test, rf_pred)))

lr_clf.fit(X_train, y_train)
lr_pred = lr_clf.predict(X_test)
print('LR정확도:{0:.4f}'.format(accuracy_score(y_test, lr_pred)))
```

- 5. 하이퍼파라미터 튜닝, 알고리즘 조절해서 좀 더 좋은 결과물을 만든다.
