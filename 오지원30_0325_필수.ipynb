# 타이타닉 데이터 전처리

```python
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
df
```

# 결측치 확인

```python
df.isna().sum()   #총 4개의 컬럼에 결측치가 존재하는 것을 확인
```

# 결측치 처리

- 결측치 삭제

```python
df.deck.value_counts()
```

```python
df
```

```python
df = df.drop(['deck'], axis=1)    # 결측치가 많은 deck열 제거
df
```

- 결측치 대체

```python
df.embarked.value_counts()
```

```python
df['embarked'].mode()   #embarked 컬럼의 최빈값 확인
```

```python
df['embarked']= df['embarked'].fillna('S')   #최빈값으로 결측치를 대체함
df
```

```python
df['embark_town'].mode() 
```

```python
df['embark_town']= df['embark_town'].fillna('Southampton')   #동일하게 최빈값으로 결측치를 대체함
df
```

```python
df['age']= df['age'].fillna(df['age'].mean())   # 나이의 전체 평균으로 결측치를 대체함
df
```

```python
df.isna().sum()  # 결측치 모두 처리 완료
```
