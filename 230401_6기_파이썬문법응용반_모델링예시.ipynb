### ML 알고리즘으로 타이타닉 생존자 예측해보기!
- 의사결정나무
- 랜덤포레스트

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# 결정트리, 랜포, 로지스틱 회귀 만들기
dt_clf = DecisionTreeClassifier(random_state=111)
rf_clf = RandomForestClassifier(random_state=111)
lr_clf = LogisticRegression(solver ='liblinear')
```

```python
dt_clf.fit(X_train, y_train)
dt_pred = dt_clf.predict(X_test)
print('DT정확도:{0:.4f}'.format(accuracy_score(y_test, dt_pred)))

rf_clf.fit(X_train, y_train)
rf_pred = rf_clf.predict(X_test)
print('RF정확도:{0:.4f}'.format(accuracy_score(y_test, rf_pred)))

lr_clf.fit(X_train, y_train)
lr_pred = lr_clf.predict(X_test)
print('LR정확도:{0:.4f}'.format(accuracy_score(y_test, lr_pred)))
```

- 1. 컬럼이 많다고해서 좋은 결과물이 나오는 건 아니다.
- 2. 사람마다 전처리하는 방법이나, 도메인 지식이 다 다르다.
- 3. 전처리 과정에서 정확도가 많은 변화가 있을 수 있다.
- 4. 응용반, 기초반 데이터 분석 전처리에 많은 시간을 투자하면서 성능을 잡아야 한다.
- 5. 하이퍼파라미터 튜닝, 알고리즘 조절해서 좀 더 좋은 결과물을 만든다.

- 7. 어떤 변수가 y값에 영향을 주는 것인가 이걸 파악하고 이 변수와 너무나 상관성이 높은 변수하게 되면 다중공선성 
- 8. 종합해서 보면 y값과 변수와의 관계를 잘 이해하고 선택해야 한다.
